```python
population_data['Notes'] = pd.to_numeric(population_data['Notes'], errors = 'coerce')
placeholder_value = -1
population_data['Notes'].fillna(placeholder_value, inplace = True)
population_data['Notes'] = population_data['Notes'].astype(int)
```

```python
sns.distplot(population_data['Notes'])
```

Thus, it is a skewed one and we cannot use mean over here

```python
population_data['Notes'].median()
```

```python
population_data['Notes'].fillna(population_data['Notes'].median(), inplace = True)
```

```python
population_data['Notes'].isnull().value_counts()
```

```python
population_data['ISO3_code'].mode()[0] # Since, it's a string we are using mode to replace missing values
```

BAR CHARTS

```python
# 1. Population Density by Location Type
density_by_type = population_data.groupby('LocTypeName')['PopDensity'].mean().sort_values()
density_by_type.plot(kind='barh', figsize=(10,6))
plt.title('Population Density by Location Type')
plt.xlabel('Population Density')
plt.ylabel('Location Type')
plt.show()
```

```python
# 2. Population Sex Ratio by Location Type
sex_ratio_by_type = population_data.groupby('LocTypeName')['PopSexRatio'].mean().sort_values()
sex_ratio_by_type.plot(kind='barh', figsize=(10,6), color='skyblue')
plt.title('Population Sex Ratio by Location Type')
plt.xlabel('Population Sex Ratio')
plt.ylabel('Location Type')
plt.show()
```

```python
# 3. Net Migration Rate by Location Type
migration_by_type = population_data.groupby('LocTypeName')['NetMigrations'].mean().sort_values()
migration_by_type.plot(kind='barh', figsize=(10,6), color='salmon')
plt.title('Net Migration Rate by Location Type')
plt.xlabel('Net Migration Rate')
plt.ylabel('Location Type')
plt.show()
```

```python
# 4. Population Growth Rate by Variant
growth_by_variant = population_data.groupby('Variant')['PopGrowthRate'].mean().sort_values()
growth_by_variant.plot(kind='bar', figsize=(10,6), color='green')
plt.title('Population Growth Rate by Variant')
plt.xlabel('Variant')
plt.ylabel('Population Growth Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Birth Rate (CBR) by Variant
birth_rate_by_variant = population_data.groupby('Variant')['CBR'].mean().sort_values()
birth_rate_by_variant.plot(kind='bar', figsize=(10,6), color='orange')
plt.title('Birth Rate (CBR) by Variant')
plt.xlabel('Variant')
plt.ylabel('Birth Rate (CBR)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 6. Median Age by Location Type
median_age_by_type = population_data.groupby('LocTypeName')['MedianAgePop'].mean().sort_values()
median_age_by_type.plot(kind='barh', figsize=(10,6), color='purple')
plt.title('Median Age by Location Type')
plt.xlabel('Median Age')
plt.ylabel('Location Type')
plt.show()
```

```python
# 7. Life Expectancy (LE) by Location Type
life_expectancy_by_type = population_data.groupby('LocTypeName')['LEx'].mean().sort_values()
life_expectancy_by_type.plot(kind='barh', figsize=(10,6), color='green')
plt.title('Life Expectancy (LE) by Location Type')
plt.xlabel('Life Expectancy (LE)')
plt.ylabel('Location Type')
plt.show()
```

```python
# 8. Under 5 Mortality Rate by Location Type
under5_mortality_by_type = population_data.groupby('LocTypeName')['Under5Deaths'].mean().sort_values()
under5_mortality_by_type.plot(kind='barh', figsize=(10,6), color='orange')
plt.title('Under 5 Mortality Rate by Location Type')
plt.xlabel('Under 5 Mortality Rate')
plt.ylabel('Location Type')
plt.show()
```

PIE CHARTS

```python
# 1. Population Distribution by Location Type
population_by_type = population_data.groupby('LocTypeName')['TPopulation1Jan'].sum().sort_values()
plt.figure(figsize=(8, 8))
plt.pie(population_by_type, labels=population_by_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution by Location Type')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# 2. Gender Distribution
male_population = population_data['TPopulationMale1July'].sum()
female_population = population_data['TPopulationFemale1July'].sum()
gender_distribution = [male_population, female_population]
gender_labels = ['Male', 'Female']
plt.figure(figsize=(8, 8))
plt.pie(gender_distribution, labels=gender_labels, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
# 3. Life Expectancy (LE) by Location Type
le_by_type = population_data.groupby('LocTypeName')['LEx'].mean().sort_values()
plt.figure(figsize=(8, 8))
plt.pie(le_by_type, labels=le_by_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Life Expectancy (LE) by Location Type')
plt.axis('equal')
plt.show()
```
